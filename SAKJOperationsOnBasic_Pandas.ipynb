Adding New Columns – Feature Engineering

```python

df["sepal_area"] = df["sepal_length"] * df["sepal_width"]
df.head()
```

```python
df["petal_length_mm"] = df["petal_length"] * 10
```

```python
df.sort_values(by="sepal_area", ascending=False).head()
```

GroupBy – Species-wise Analysis

```python
df.groupby("species")["petal_length"].mean()
```

```python
df.groupby("species").agg({
    "sepal_length": ["mean", "max"],
    "petal_length": ["mean", "min"],
    "petal_width": "mean"
})
```
